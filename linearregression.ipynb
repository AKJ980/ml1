```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("linear_Salary_Data.csv")
X, y = df[["YearsExperience"]], df["Salary"]
m = LinearRegression().fit(X, y); yp = m.predict(X)

plt.scatter(X, y); plt.plot(X, yp, color='red'); plt.show()

cost = np.mean((y - yp) ** 2)
SSE = np.sum((y - yp) ** 2)
SSR = np.sum((yp - np.mean(y)) ** 2)
SST = np.sum((y - np.mean(y)) ** 2)
R2 = SSR / SST
print(f"Cost (MSE): {cost:.4f}")
print(f"SSE: {SSE}, SSR: {SSR:.4f}, SST: {SST:.4f}, R²: {R2:.4f}")
```
